```python
import pandas as pd
```

```python
df =pd.read_csv('HRDataset_v14.csv',sep=',')
```

```python
df_org = pd.DataFrame(columns=['nomeColaborador',
                            'isCasado',
                            'genero',
                            'isProgramaDiversidade',
                            'salario',
                            'cargo',
                            'estado',
                            'zipCode',
                            'sexo',
                            'dataContratacao',
                            'dataDesligamento',
                            'statusColaborador',
                            'departamento',
                            'fonteRecrutamento',
                            'performanceScore',
                            'engajamento',
                            'satisfacao',
                            'ultimaDataAvaliacaoPerformance',
                            'faltas'])
i=1
for index, row in df.iterrows():
    if row['MarriedID']==1:
      casamento='Casado(a)'
    else:
      casamento='Solteiro(a)'
    if row['GenderID']==1:
      genero='Masculino'
    else:
      genero='Feminino'  
    if row['FromDiversityJobFairID']==1:
      diversidade='Sim'
    else:
      diversidade='Não'
    line=row['DateofHire']  
    if line.find('/')==1:
       line='0'+line
       mes=line[0:2]
    else:
       mes=line[0:2]
    line=line[3:]   
    if line.find('/')==1:
       line='0'+line
       dia=line[0:2]
    else:
       dia=line[0:2]
    line=line[3:]
    ano=line
    str_convert =str(row['DateofTermination'])
    if str_convert == 'nan':
      datacomp= 'null'
    else:  
      lineTerm=row['DateofTermination']   
      if lineTerm.find('/')==1:
         lineTerm='0'+lineTerm
         mes2=lineTerm[0:2]
      else:
         mes=lineTerm[0:2]
      lineTerm=lineTerm[3:]   
      if lineTerm.find('/')==1:
         lineTerm='0'+lineTerm
         dia2=lineTerm[0:2]
      else:
         dia2=lineTerm[0:2]
      lineTerm=lineTerm[3:]
      ano2=lineTerm 
      datacomp=dia2+'/'+mes2+'/'+ano2       
    primeiro_nome= row['Employee_Name'][row['Employee_Name'].find(',')+2:len(row['Employee_Name'])]
    segundo_nome= row['Employee_Name'][0:row['Employee_Name'].find(',')]
    nome_completo= primeiro_nome+' '+segundo_nome        
    df_org = df_org.append(pd.DataFrame({'nomeColaborador':nome_completo,
                          'isCasado':casamento,
                          'genero':genero,  
                          'isProgramaDiversidade':diversidade,                     
                          'salario':row['Salary']/12,
                          'cargo':row['Position'],
                          'estado':row['State'],
                          'zipCode':row['Zip'],
                          'sexo':row['Sex'],
                          'dataContratacao':dia+'/'+mes+'/'+ano,
                          'dataDesligamento':datacomp,
                          'statusColaborador':row['EmploymentStatus'],
                          'departamento':row['Department'],
                          'fonteRecrutamento':row['RecruitmentSource'],
                          'performanceScore':row['PerformanceScore'],
                          'engajamento':row['EngagementSurvey'],
                          'satisfacao':row['EmpSatisfaction'],
                          'ultimaDataAvaliacaoPerformance':row['LastPerformanceReview_Date'],
                          'faltas':row['Absences']}, index=[i]))
df_org=df_org.reset_index() 
df_org=df_org.drop(columns=['index'])
df_org
```

```python
df_org.to_csv('datasetRH.csv')
```
